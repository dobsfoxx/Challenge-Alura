### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head(10)
```

#Análise de Faturamento

```python
faturamento = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()
```

```python
import matplotlib.pyplot as plt

faturamentos = [faturamento, faturamento2, faturamento3, faturamento4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamentos, color=['blue', 'orange', 'green', 'red'], width=0.3)
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.title('Faturamento por Loja')
plt.text(0, faturamento, f'R${faturamento:.2f}', ha='center', va='bottom')
plt.text(1, faturamento2, f'R${faturamento2:.2f}', ha='center', va='bottom')
plt.text(2, faturamento3, f'R${faturamento3:.2f}', ha='center', va='bottom')
plt.text(3, faturamento4, f'R${faturamento4:.2f}', ha='center', va='bottom')
plt.show()
```

# 2. Vendas por Categoria

```python
vendas_loja = loja.value_counts('Categoria do Produto')
```

```python
vendas_loja2 = loja2.value_counts('Categoria do Produto')
```

```python
vendas_loja3 = loja3.value_counts('Categoria do Produto')
```

```python
vendas_loja4 = loja4.value_counts('Categoria do Produto')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
vendas = [vendas_loja, vendas_loja2, vendas_loja3, vendas_loja4]

plt.figure(figsize=(20, 8))
for i in range(4):
  plt.subplot(2, 2, i + 1)
  categorias_ordenadas = vendas[i].sort_index()
  labels = [f'{cat}: {count}' for cat, count in categorias_ordenadas.items()]


  cores = plt.get_cmap('tab10').colors
  plt.pie(categorias_ordenadas.values, labels=labels, textprops={'fontsize': 9}, colors=cores)
  plt.title(f'Vendas por Categoria - {lojas[i]}', fontsize=14)
  plt.ylabel('')
  plt.xticks([])
  plt.yticks([])


plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_mais_vendidos_loja = loja.value_counts('Produto').sort_values(ascending=False)
mais_vendidos = produtos_mais_vendidos_loja.head(5)
menos_vendidos = produtos_mais_vendidos_loja.tail(5)
produtos_mais_vendidos_loja

```

```python
produtos_mais_vendidos_loja2 = loja2.value_counts('Produto').sort_values(ascending=False)
mais_vendidos_2 = produtos_mais_vendidos_loja2.head(5)
menos_vendidos_2 = produtos_mais_vendidos_loja2.tail(5)
produtos_mais_vendidos_loja2
```

```python
produtos_mais_vendidos_loja3 = loja3.value_counts('Produto').sort_values(ascending=False)
mais_vendidos_3 = produtos_mais_vendidos_loja3.head(5)
menos_vendidos_3 = produtos_mais_vendidos_loja3.tail(5)

produtos_mais_vendidos_loja3


```

```python
produtos_mais_vendidos_loja4 = loja4.value_counts('Produto').sort_values(ascending=False)
mais_vendidos_4 = produtos_mais_vendidos_loja4.head(5)
menos_vendidos_4 = produtos_mais_vendidos_loja4.tail(5)

produtos_mais_vendidos_loja4


```

```python
mais_vendidos_lojas = [mais_vendidos, mais_vendidos_2, mais_vendidos_3, mais_vendidos_4]
menos_vendidos_lojas = [menos_vendidos, menos_vendidos_2, menos_vendidos_3, menos_vendidos_4]

plt.figure(figsize=(15, 8))

for i in range(4):
  plt.subplot(2, 2, i + 1)
  plt.barh(mais_vendidos_lojas[i].index, mais_vendidos_lojas[i].values, color='green', label='Mais Vendidos')
  plt.barh(menos_vendidos_lojas[i].index, menos_vendidos_lojas[i].values, color='red', label='Menos Vendidos')
  plt.xlabel('Quantidade Vendida')
  plt.ylabel('Produto')
  plt.title(f'Produtos Mais e Menos Vendidos - Loja {i + 1}')
  plt.grid(axis='x', linestyle='--', alpha=0.7)
  plt.legend()

plt.tight_layout()
plt.show()
```
